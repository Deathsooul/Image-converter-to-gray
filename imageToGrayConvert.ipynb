```
import matplotlib.image as mpimg
import cv2
import os, sys
from PIL import Image
import glob

from google.colab import drive
drive.mount('/content/drive')
```

```
path = '/content/drive/My Drive/Faculdade/2020/InteligenciaArtificial/2Bim/imageToGrayConvert/images'
pathTo = '/content/drive/My Drive/Faculdade/2020/InteligenciaArtificial/2Bim/imageToGrayConvert/images-grey/'

for filename in glob.glob(path + '/*.jpg'):       
    imgName=os.path.basename(filename)
    image = cv2.imread(filename)
    image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    cv2.imwrite(pathTo+imgName, image_gray)
      
```
